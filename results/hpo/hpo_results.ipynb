```python
from graphdriver.commons import results, config, setup
import hpbandster.core.result as hpres
from graphdriver.utils import paths
import numpy as np
```

```python
cancer = 'brca'
network_type = ["ppi"]
outer_fold = 0
```

```python
for c in setup.cancer_types():
    print(c)
    for nt  in setup.network_types()[-1:]:
        scores = []
        for outer_fold in range(10):
            path = paths.results_hpo(cancer=c, network_type=nt, outer_fold=outer_fold)
            res = hpres.logged_results_to_HBS_result(path)
            res_budget = res.get_all_runs()
            res_budget = [r for r in res_budget if r.budget == 4]
            res_budget.sort(key=lambda x: x['loss'], reverse=False)
            scores.append(res_budget[0].info["pr_auc_mean_test"])
        print(nt)
        print(np.mean(scores))
        print(np.std(scores))
    print("")
```

```python
for c in setup.cancer_types():
    print(c)
    for nt in setup.network_types():
        print(nt)
        res = results.load_results(cancer=c, network_type=nt)
        print(res.score_test_pr_auc())
    print()
        
```
