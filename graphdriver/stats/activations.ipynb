```python
neg_range = 2
pos_range = 2
spacing = 1000
x = np.linspace(0,pos_range+neg_range,spacing*(pos_range + neg_range)) - neg_range
# x = np.arange(neg_range + pos_range) - neg_range
y_relu = np.concatenate((np.zeros(neg_range*spacing),np.arange(pos_range*spacing)/spacing))
y_leaky_relu = np.concatenate(((np.arange(neg_range*spacing)/spacing-neg_range)*0.1,np.arange(pos_range*spacing)/spacing))
y_gelu = 0.5 * x * (1+np.tanh(np.sqrt(2/np.pi)*(x+0.044715*np.power(x,3))))
plt.plot(x,y_leaky_relu, color='orange')
plt.plot(x,y_relu, color='blue')
plt.plot(x,y_gelu, color='green')
plt.show()
# https://paperswithcode.com/method/gelu
```

```python
import sklearn.metrics as m
import matplotlib.pyplot as plt
import numpy as np
```

```python
negs = 2000
pos = 20
true_hat = np.random.uniform(low=0.0, high=1.0, size=pos)+0.2
true_hat = np.clip(true_hat,0.0,1.)
true = np.ones((pos))
false_hat = np.random.uniform(low=0.0, high=1.0, size=negs)-0.2
false_hat = np.clip(false_hat,0.,1.)
false = np.zeros((negs))
y_score = np.concatenate((true_hat,false_hat))
y_true = np.concatenate((true,false))

t,f, _ =m.roc_curve(y_true, y_score)
plt.plot(f,t)
precision, recall, thresholds =m.precision_recall_curve(y_true, y_score)
plt.plot(recall,precision)
plt.show()
```
