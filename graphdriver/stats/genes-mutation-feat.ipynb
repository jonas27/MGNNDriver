```python
from graphdriver.commons import data, setup
import torch
import matplotlib.pyplot as plt
```

```python
for cancer in setup.cancer_types():
    d = data.Dataset(cancer).get_data()
    drivers = torch.where(d.y == 1)[0]
    passengers = torch.where(d.y == 0)[0]
    candidates = torch.where(d.y == -1)[0]
    drivers_mean = d.x[drivers].mean(dim=0)
    passengers_mean = d.x[passengers].mean(dim=0)
    candidates_mean = d.x[candidates].mean(dim=0)
    plt.plot(torch.arange(12), drivers_mean, label='drivers')
    plt.plot(torch.arange(12), passengers_mean, label='passengers')
    plt.plot(torch.arange(12), candidates_mean, label='candidates')
    plt.legend()
    plt.show()
    plt.close()
```
