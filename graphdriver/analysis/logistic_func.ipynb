```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))
z_deriv = z * (1 - z )
  
plt.plot(x, z, label='Logistic function')
plt.plot(x, z_deriv, label='Logistic derivative')
plt.xlabel("x")
plt.ylabel("Logistic(x)")
plt.legend(loc="upper left")
  
plt.show()
```
